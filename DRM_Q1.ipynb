```python
import pandas as pd
import numpy as np
```

```python
df_banknifty = pd.read_csv("Q1 - BANKNIFTY FUTURES.csv")
df_currency = pd.read_csv("Q1 - EUROINR FUTURES.csv")
```

```python
df_banknifty = df_banknifty[:16]
df_currency = df_currency[:18]
```

```python
cci = [0.08, 0.065] # Continuously compounded annual domestic borrowing and lending interest rates [Borrowing, Lending]
ccfi = [0.07, 0.05] # Continuously compounded annual foreign borrowing and lending interest rates [Borrowing, Lending]
```

```python
#For our case we take dividend yeild = 0
```

```python
# We define a fuction for evaulating the arbitrage opportunity depending on the spot ask and bid we calculates the bid and ask for futures.
# On comparing this with the Futures bid, if futures bid is greater then we short futures
# If Calculated Ask is greater than Futures ask, then we are long on futures.If noth are equal then there is no arbitrage opportunity
def evaluate_arbitrage(cci: [float, float], ccfi: [float, float], spot_prices: [float, float], futures_prices: [float, float],time_to_maturity: float, type: int):
    domestic_borrowing_rate, domestic_lending_rate = cci
    foreign_borrowing_rate, foreign_lending_rate = ccfi
    spot_bid, spot_ask = spot_prices
    futures_bid, futures_ask = futures_prices

    if type == 0:
        yield_rates = [0, 0]
    elif type == 1:
        yield_rates = [foreign_borrowing_rate, foreign_lending_rate]
    else:
        raise ValueError('Type must be either 0 for stocks futures or 1 for currency futures')

    calculated_bid = spot_ask * np.exp((domestic_borrowing_rate - yield_rates[1]) * time_to_maturity)
    calculated_ask = spot_bid * np.exp((domestic_lending_rate - yield_rates[0]) * time_to_maturity)

    if calculated_bid < futures_bid:
        return "Short on Futures"
    if calculated_ask > futures_ask:
        return "Long on Futures"
    else:
        return "No Arbitrage Opportunity"
```

```python
df_banknifty["Arbitrage"]=np.nan
for index, row in df_banknifty.iterrows():
    spot_price = [row['Spot Bid'],row['Spot Ask']]
    futures_price = [row['Futures Bid'],row['Futures Ask']]
    ttm = row['TTM']
    df_banknifty['Arbitrage'][index] = evaluate_arbitrage(cci,ccfi,spot_price,futures_price,ttm,0)
```
